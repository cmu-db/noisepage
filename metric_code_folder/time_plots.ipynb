```python
import pandas as pd
import matplotlib.pyplot as plt
from matplotlib.ticker import MaxNLocator
from matplotlib.figure import figaspect
```

```python
# 8 threads
df = pd.read_csv ('daf_time_agg_woLogging_8thread.csv', header=0, skipinitialspace=True)
# for col in df.columns: 
#     print(col)

# draw txn_removal, unlink, and num_txn
# plt.xticks(fontsize=8)
# plt.yticks(fontsize=8)
w, h = figaspect(.2)

fig, axes = plt.subplots(nrows=1, ncols=2, figsize=(w, h))

axes[0].xaxis.set_major_locator(MaxNLocator(nbins=15))
axes[0].yaxis.set_major_locator(MaxNLocator(nbins=8))

axes[1].xaxis.set_major_locator(MaxNLocator(nbins=15))
axes[1].yaxis.set_major_locator(MaxNLocator(nbins=8))

axes[0].plot( 'start_time', 'TXN_REMOVAL', data=df, marker='', color='royalblue', linewidth=1)
axes[0].plot( 'start_time', 'UNLINK', data=df, marker='', color='olive', linewidth=1)
axes[1].plot( 'start_time', 'total_num_txns', data=df, marker='', color='red', linewidth=1.5)
fig.tight_layout()
fig.legend()
fig.savefig('time_actions_8thread.eps', format='eps')
```
